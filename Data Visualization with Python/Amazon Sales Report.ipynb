## Exploratory Data Analysis (EDA)

### Size Column for EDA 

```python
#using Seaborn 

ax=sns.countplot(x='Size',data=df)
#as we can see we cannot see the labels properly or the values at which the data is distributed, hence we need a for loop to properly label it 
```

```python
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

## Observation: we can see that most of the customers buy M-size 

# Group By 
### a function in pandas that is used to group more than one columns in a dataframe

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
#to put it in graph
Qty_s=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
sns.barplot(x='Size', y='Qty',data=Qty_s)
```

## Observation: We can now clearly see in order which size is most bought by customers

### Courier Status Column for EDA

```python
plt.figure(figsize=(10,5))
ax1=sns.countplot(x='Courier Status',data=df, hue='Status')
plt.show()
```

## Observation: Majority of the products/orders have been succesfully delivered to the buyer

```python
#To create histogram
df['Size'].hist()
```

```python
df['Category'].hist()
```

## Observation: The histogram is not clearly understandable because there is no spcific boundary for each category

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

## Observation: Clear Boundaries for each category and understandable . Also, we can observe that most orders are for T-shirt

# B2B column for EDA 

```python
#Pie Chart
B2B_Check=df['B2B'].value_counts()
plt.pie(B2B_Check,labels=B2B_Check.index, autopct= '%1.1f%%')
plt.axis('equal')
plt.show()
```

## Observation: False means that maximum buyers (99.2%) are retailers and 0.8% are B2B

```python
#Scatter Plot
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

## Observation: We can check for each category which size is available

# Cities column for EDA

```python
#plot count
plt.figure(figsize=(12,7.5))
sns.countplot(x='ship-state',data=df)
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution within the states')
plt.xticks(rotation=90)
plt.show()
```

```python
#to obtain graph for first 10 only
top10=df['ship-state'].value_counts().head(10)
#plot count
plt.figure(figsize=(12,7.5))
sns.countplot(x='ship-state',data=df[df['ship-state'].isin(top10.index)])
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution within the states')
plt.xticks(rotation=90)
plt.show()
```

## Observation: We can see that most of the buyers are from Maharashtra
